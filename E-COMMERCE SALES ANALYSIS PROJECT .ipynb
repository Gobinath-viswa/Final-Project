##Stage 2 – Data Cleaning and Pre-processing
Handle missing values (impute or drop)

Handle duplicates

Convert data types if needed

### Handling Missing Data

```python
# Check Missing Values
df.isnull().sum()
```

```python
df['Sales'].fillna(df['Sales'].mean())
df['Profit'].fillna(df['Profit'].mean())
```

```python
# Categorical columns – fill with mode
df['Category'].fillna(df['Category'].mode()[0])
df['Payment Mode'].fillna(df['Payment Mode'].mode()[0])
```

### Handling Duplicate Records

```python
## Handling Duplicate Values
df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

### Data Formatting & Type Conversion

```python
# Convert Order Date to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
# Standardize column names
df.columns = df.columns.str.lower().str.replace(" ", "_")
```

### Handling Incorrect / Invalid Values

```python
# Remove negative or zero sales
df = df[df['sales'] > 0]
```

```python
# Ensure quantity is positive
df = df[df['quantity'] > 0]
```

### Feature Engineering (Derived Features)

```python
# Total price before discount
df['gross_sales'] = df['quantity'] * df['unit_price']
```

```python
# Profit margin
df['profit_margin'] = (df['profit'] / df['sales']) * 100
```

```python
# Month extraction
df['order_month'] = df['order_date'].dt.month
```

### Data Transformation

```python
## Data Transformation
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[['sales', 'profit']] = scaler.fit_transform(df[['sales', 'profit']])
```

```python
## Encoding Categorical Variables
df_encoded = pd.get_dummies(df, columns=['category', 'region', 'payment_mode'])
```

```python
## Cleaned Dataset Check
df_encoded.info()
df_encoded.head()
```

## **Stage 3 – EDA and Visualizations**

Univariate Analysis → distribution of single variables (countplot, histogram, boxplot)

Bivariate Analysis → relation between two variables (scatterplot, barplot, correlation heatmap)

Multivariate Analysis → relation among 3+ variables (pairplot, grouped analysis, pivot tables, advanced plots)

## **COUNT PLOT - ORDERS PER CATEGORY**

```python
import seaborn as sns
plt.figure(figsize=(8, 6))
sns.countplot(y='category', data=df)
plt.title("Order Count by Category")
plt.show()
```

This chart shows the number of orders placed in each product category.
It helps identify customer demand across categories.
Categories such as Books, Kitchen, Furniture, and Home Decor receive more orders, indicating higher popularity.

##**HISTOGRAM - SALES AMOUNT DISTRIBUTION**

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(df['sales'], bins=20)
plt.title("Distribution of Sales")
plt.xlabel("Sales Amount")
plt.ylabel("Frequency")
plt.show()
```

The histogram displays how sales values are distributed.
Most orders fall in the low to mid sales range, while fewer orders generate very high sales.
This indicates a right-skewed distribution, common in e-commerce data.

# **BOX PLOT - SALES OUTLIERS**

```python
plt.figure(figsize=(6, 6))
sns.boxplot(
    y='sales',
    data=df,
    boxprops={'facecolor': 'green', 'edgecolor': 'black', 'alpha': 0.5},
    whiskerprops={'color': 'black', 'linewidth': 2},
    capprops={'color': 'green'},
    flierprops={'marker': 'o', 'markerfacecolor':'red', 'markeredgecolor':'black'},

)
plt.title("Sales Outliers")
plt.ylabel("Sales Amount")
plt.show()
```

This chart highlights the median, spread, and outliers in sales.
Several high-value outliers are visible, representing bulk purchases or premium products.
Most transactions lie within a consistent sales range.

## **SCATTER PLOT - SALES VS QUANTITY**

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df,
    x='quantity',
    y='sales',
    hue='category',
    palette='Greens')
plt.title("Sales vs Quantity")
plt.legend(
    title="Category",
    loc='upper left',
    bbox_to_anchor=(1, 0.5))
plt.xlabel("Quantity")
plt.ylabel("Sales")
plt.show()
```

The scatter plot shows the relationship between quantity sold and sales value.
A positive relationship is observed—higher quantities generally result in higher sales.
Fixed quantity levels create vertical patterns.

# **BAR CHART - AVERAGE SALES BY CATEGORY**

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df,
    x='category',
    y='sales',
    estimator='mean',
    hue='category',
    palette='viridis')

plt.title("Average Sales by Category")
plt.ylabel("Average Sales")
plt.show()
```

This chart compares the average sales value across categories.
Categories like Electronics, Home Decor, and Beauty have higher average sales.
Even categories with fewer orders can contribute significant revenue.

## **HEATMAP - CORRELATION MATRIX**

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df[['sales','profit','quantity','discount']].corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

The heatmap shows relationships between numerical variables.

Sales and Profit have a strong positive correlation

Quantity and Sales show moderate correlation

Discount and Profit show weak or negative correlation

This indicates that excessive discounts may reduce profitability

## **GROUPED BAR CHART - CATEGORY & REGION SALES**

```python
plt.figure(figsize=(10, 6))
pd.pivot_table(df, values='sales', index='category', columns='region', aggfunc='sum').plot(kind='bar')
plt.title("Category vs Region Sales")
plt.show()
```

This chart compares total sales by category across regions.
The North and West regions perform strongly across most categories.
Regional sales patterns differ depending on product category.

## **BOX PLOT - SALES BY REGION**

```python
sns.boxplot(x='region', y='sales', data=df)
plt.title("Sales Distribution by Region")
plt.show()
```

This chart compares sales distribution across regions.
Median sales are similar across regions, but North and West show more high-value outliers.
This indicates larger orders in those regions.

## **PIE CHART – PAYMENT METHOD SHARE**

```python
df['payment_mode'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("Payment Method Distribution")
plt.ylabel("")
plt.show()
```

The pie chart shows the percentage distribution of payment methods.
UPI, Credit Card, Debit Card, COD, and Net Banking are almost equally used.
This shows customers prefer multiple flexible payment options.

## **LINE CHART - SALES TREND OVER TIM**

```python
plt.figure(figsize=(10, 6))
df.groupby('order_date')['sales'].sum().plot()
plt.title("Sales Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.show()
```

This chart displays monthly sales trends.
Sales remain stable with periodic peaks, likely due to promotions or seasonal demand.
Overall performance is consistent across time.

## **BUBBLE CHART - SALES VS PROFIT VS QUANTITY**

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['sales'], df['profit'], s=df['quantity']*10)
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.title("Sales vs Profit vs Quantity")
plt.show()
```

This chart represents three variables together:

X-axis: Sales

Y-axis: Profit

Bubble size: Quantity

Higher sales generally result in higher profit, and larger bubbles show bulk purchases generating more revenue. Some high-sales orders yield lower profit due to discounts.
